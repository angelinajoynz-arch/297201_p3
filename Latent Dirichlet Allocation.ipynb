### LatentDirichletAllocation()

A 'probabilistic generative algorithm' used for topic modeling in natural language processing.

Gives us 'topics' -  A ranked list of the words that contribute most heavily to each topic's definition. The words themselves do kind of need human interpretation as to what the 'theme' might be, though, and actually label them. LDA don't do that part, it just groups words together.

Scikit-learn `sklearn`
- `LatentDirichletAllocation`: natural language processing/learning model for topic modelling
- `CountVectorizer` for word tokenisation (splitting larger strings into individual words), needed for LDA


Natural Language Toolkit `nltk`
- includes expanded list of stop-words (for use with LDA, includes context words like 'is', 'and', 'they', etc that don't have any real meaning)

```python
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
poems = pd.read_excel('PoetryforNLP.xlsx')
```

```python
# Load the default English stop words
stop_words = stopwords.words('english')
```

Note: the "stop words"  list that comes from the nltk package is already pretty exhaustive but if there are words we are looking to actively ignore the list can be added to by adding these lines:


`stop_words.extend(['example', 'words'])`

`stop_words = list(stop_words)`

```python
these_poems = poems.Poem

# Vectorize: Convert text to a Count Matrix for use with LDA (removing stop words)
vectorizer = CountVectorizer(stop_words=stop_words)
poem_term_matrix = vectorizer.fit_transform(these_poems)
```

Note: `n_components` = the number of topics you would like it to output

```python
# Model: Fit LDA (identifying 5 topics, currently)
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(poem_term_matrix)

# Get the words out
words = vectorizer.get_feature_names_out()
```

```python
# Print top words for each topic
words = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"\nTopic #{topic_idx}:")
    print([words[i] for i in topic.argsort()[:-6:-1]])
```

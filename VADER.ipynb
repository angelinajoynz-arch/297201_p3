### VADER

VADER (Valence Aware Dictionary and sEntiment Reasoner) is a sentiment analysis tool that takes a "bag of words" rule-based approach to analyse the text it's presented with, and then outputs a score (between -1 and 1) for 3 sentiment attributes - 'positive', 'negative' and 'neutral'. It then also gives a fourth 'compound' score, which basically mashes all of those 3 attributes into one combined overall score.

Note: I have checked multiple times to be absolutely sure, too - VADER is not an LLM, so there's no risk of accidentally using AI for this project when we're not supposed to. There were other Python tools I was looking at that did something similar, and were slightly better at it, but they're either LLMs or LLM-adjacent.

It can be installed with the command `pip install vaderSentiment`

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Initialize the analyzer
analyzer = SentimentIntensityAnalyzer()
```

This first test should give a negative score, because it's a negative statement.

```python
text1 = "I greatly dislike rice risotto. It tastes like cardboard and regret."

scores_text1 = analyzer.polarity_scores(text1)

print(scores_text1)
```

This next one should result in a more positive score, because it has a positive sentiment.

```python
text_to_analyse = "I am very much a dog person. I love dogs, they're the absolute best things ever."

scores = analyzer.polarity_scores(text_to_analyse)

print(scores)
```

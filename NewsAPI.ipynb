Import the API Package

```python
from newsapi import NewsApiClient
```

Insert your API key in order to access the API from Python (I can't share mine cos I think that's technically a breach of their ToS)

```python
your_API_Key = 'Replace this string with your API key'
newsapi = NewsApiClient(api_key=your_API_Key)
```

get_everything() gets <em>everything</em> specified to it - the NewsAPI website has full documentation for [Everything](https://newsapi.org/docs/endpoints/everything), but these are a few options:
- `q` : query
- `domains` : the URLs for specific sources (formatted as 'bbc.co.uk')
- `from_param`, `to` : dates to search between (format: yyyy MM dd)
- `language` : the language the article is in - probably best to select `en` for English

<strong>Limitation</strong>: `get_everything` doesn't allow us to search by country, only sources

```python
response = newsapi.get_everything(q='Climate Change', language='en', from_param='2026-05-04', to='2026-05-04')
```

```python
response # returns a dict object
```

```python
response['articles'] # returns a list
```

```python
# To see all the titles of the articles in that list...
articles = response['articles']
for i in articles:
    if 'climate' in i['title']:
        print(f"Title: {i['title']}")
```
